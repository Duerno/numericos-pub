```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dados de entrada
ts = np.array(
     [   0.        ,    5.26315789,   10.52631579,   15.78947368,
        21.05263158,   26.31578947,   31.57894737,   36.84210526,
        42.10526316,   47.36842105,   52.63157895,   57.89473684,
        63.15789474,   68.42105263,   73.68421053,   78.94736842,
        84.21052632,   89.47368421,   94.73684211,  100.        ]
)

ys = np.array(
     [ 0.2493938 ,  0.72091586,  1.08935177,  0.82557285,  0.15184373,
      -0.55819513, -0.84832015, -1.13075724, -0.66132559, -0.37482053,
       0.49910671,  0.99533713,  0.97493417,  0.56770796, -0.59740919,
      -0.76333201, -1.05362856, -0.8085086 , -0.32029816,  0.4021104 ]
)
```

```python
# Mostra o conjunto de dados
plt.plot(ts, ys, 'ko')
plt.show()
```

```python
def EQ(w):
    return np.sum((np.sin(w * ts) - ys)**2)

def EQ_(w):
    return np.sum(
        2 * (np.sin(w*ts) - ys) 
          * np.cos(w*ts) * ts
    )

def EQ__(w):
    e = 1e-6
    return (EQ_(w + e) - EQ_(w)) / e
```

```python
# Fazemos vários chutes do valor inicial...
w = 0.14
for i in range(30):
    w = w - EQ_(w) / EQ__(w)

    
ts2 = np.linspace(0, 100, 100)
Y = np.sin(w * ts2)
plt.plot(ts2, Y, label='teórico')
plt.plot(ts, ys, 'k+', label='dados')
plt.show()

print(w, EQ(w))

```
