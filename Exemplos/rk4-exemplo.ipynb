```python
# Dinamica gravitacional
GM = 1.5
alpha = 1.4

def f(v, t):
    x, y, vx, vy = v
    
    # Implementa o sistema de equações predador-presa
    x_ = vx
    y_ = vy
    vx_ = - GM * x / (x**2 + y**2)**(alpha)
    vy_ = - GM * y / (x**2 + y**2)**(alpha)
    
    return np.array([x_, y_, vx_, vy_])

v0 = np.array([1, 1, 1, -1])
```

```python
# Executa o método de Runge-Kutta 4a ordem
print('calculando...')
T = 500
dt = 0.1

vs = [v0]
ts = [0]

t = 0
v = v0
while t < T:
    k1 = f(v, t)
    k2 = f(v + 0.5 * dt * k1, t + 0.5 * dt)
    k3 = f(v + 0.5 * dt * k2, t + 0.5 * dt)
    k4 = f(v + 1.0 * dt * k3, t + 1.0 * dt)
    k_med = (k1 + 2*k2 + 2*k3 + k4) / 6
    v = v + k_med * dt
    t = t + dt
    
    vs.append(v)
    ts.append(t)
    
vs = np.array(vs)
ts = np.array(ts)

time.sleep(1.0)
print('ok!')
```
